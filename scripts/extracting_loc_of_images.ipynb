```python
from os import listdir
import cv2
import sys
import pandas as pd
import os
import numpy as np
from matplotlib import pyplot as plt

input_path = '../data/Set1Part0'
output_path = "pikles"
img_size=(30,30,3)


#Read in image files
# images_files = [f for f in listdir(input_path + 'dataset')]
images_files = [f for f in listdir(input_path)]

images_files = images_files[:20]

images = {}

for image_file in images_files:
    images[image_file] = (cv2.imread(input_path + "/" + image_file))
```

```python
img1 = (images[images_files[0]])
RGB_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (20,10))
plt.imshow(RGB_img, aspect="auto")
```

```python
def blue_mask(img):
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mask1 = cv2.inRange(hsv, (0, 70, 50), (10, 255, 255))
    
    cnt, _ = cv2.findContours(mask1, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
    cv2.drawContours(mask1, cnt, -1, (0,255,0), 3)
    
    return mask1
    

def red_mask(img):
    
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    mask2 = cv2.inRange(hsv, (70, 0, 0), (255, 255, 255))
    
    cnt, _ = cv2.findContours(mask2, mode=cv2.RETR_EXTERNAL, method=cv2.CHAIN_APPROX_NONE)
    cv2.drawContours(mask2, cnt, -1, (0,255,0), 3)
    
    return mask2
```

```python
blue = blue_mask(RGB_img)
plt.imshow(blue)
```

```python
red = red_mask(RGB_img)
plt.imshow(red)
```
